1. Bubble Sort
</div>

**Algorithm**

1. Bubble sort works by **repeatedly comparing adjacent elements** in a list or array.
2. If two neighboring elements are in the **wrong order**, they are **swapped**.
3. During each full pass through the array, the **largest unsorted element moves to the end** of the list, similar to a bubble rising to the top.
4. After each pass, the **last element is already sorted**, so the next pass ignores it.
5. This process continues for the remaining unsorted part of the array until **no more swaps are needed** or all passes are completed.

**Time Complexity (TC)**

* Best Case: **O(n)** (when the array is already sorted and an optimization stops early)
* Average Case: **O(n²)**
* Worst Case: **O(n²)**

**Space Complexity (SC)**

* **O(1)** (Bubble sort is an **in-place algorithm**, meaning it does not require extra memory beyond a few variables).


```python
def bubble_sort(arr):
    n = len(arr)
    snapshots = [arr.copy()]  # store initial state

    for i in range(n):
        for j in range(0, n-i-1):  
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                snapshots.append(arr.copy())  # store after every swap

    return arr, snapshots
```

```python
def visualize_sort(snapshots):
    for step, arr in enumerate(snapshots):
        clear_output(wait=True)
        plt.figure()
        plt.bar(range(len(arr)), arr)
        plt.title(f"Bubble Sort Step {step}")
        plt.xlabel("Index")
        plt.ylabel("Value")
        plt.show()
        time.sleep(0.1)


arr = [64, 34, 25, 12, 22, 11, 90]

sorted_arr, snapshots = bubble_sort(arr)

#visualize_sort(snapshots)

print("Sorted Array:", sorted_arr)
```

```python
def insertion_sort(arr):
    snapshots = [arr.copy()]  # store initial state

    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
            snapshots.append(arr.copy())  # store after every shift
        arr[j + 1] = key
        snapshots.append(arr.copy())  # store after placing the key

    return arr, snapshots
```

```python
def visualize_sort(snapshots):
    for step, arr in enumerate(snapshots):
        clear_output(wait=True)
        plt.figure()
        plt.bar(range(len(arr)), arr)
        plt.title(f"Bubble Sort Step {step}")
        plt.xlabel("Index")
        plt.ylabel("Value")
        plt.show()
        time.sleep(0.2)


arr = [64, 34, 25, 12, 22, 11, 90]

sorted_arr, snapshots = insertion_sort(arr)

#visualize_sort(snapshots)

print("Sorted Array:", sorted_arr)
```

3. Selection Sort
</div>

```python
def selection_sort(arr):
    snapshots = [arr.copy()]  # store initial state

    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
        snapshots.append(arr.copy())  # store after every swap

    return arr, snapshots
```

```python
def visualize_sort(snapshots):
    for step, arr in enumerate(snapshots):
        clear_output(wait=True)
        plt.figure()
        plt.bar(range(len(arr)), arr)
        plt.title(f"Bubble Sort Step {step}")
        plt.xlabel("Index")
        plt.ylabel("Value")
        plt.show()
        time.sleep(0.2)


arr = [64, 34, 25, 12, 22, 11, 90]

sorted_arr, snapshots = insertion_sort(arr)

#visualize_sort(snapshots)

print("Sorted Array:", sorted_arr)
```

4. Merge Sort
</div>

```python
def merge_sort(arr):
    snapshots = [arr.copy()]  # store initial state

    if len(arr) > 1:
        mid = len(arr) // 2
        L = arr[:mid]
        R = arr[mid:]

        merge_sort(L)
        merge_sort(R)

        i = j = k = 0

        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1
            snapshots.append(arr.copy())  # store after every merge step

        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1
            snapshots.append(arr.copy())  # store after every merge step

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
            snapshots.append(arr.copy())  # store after every merge step

    return arr, snapshots
```

```python
def visualize_sort(snapshots):
    for step, arr in enumerate(snapshots):
        clear_output(wait=True)
        plt.figure()
        plt.bar(range(len(arr)), arr)
        plt.title(f"Bubble Sort Step {step}")
        plt.xlabel("Index")
        plt.ylabel("Value")
        plt.show()
        time.sleep(0.2)


arr = [64, 34, 25, 12, 22, 11, 90]

sorted_arr, snapshots = insertion_sort(arr)

#visualize_sort(snapshots)

print("Sorted Array:", sorted_arr)
```

5. Quick Sort
</div>

```python
def quick_sort(arr):
    snapshots = [arr.copy()]  # store initial state

    if len(arr) <= 1:
        return arr, snapshots
    else:
        pivot = arr[len(arr) // 2]
        left = [x for x in arr if x < pivot]
        middle = [x for x in arr if x == pivot]
        right = [x for x in arr if x > pivot]
        sorted_arr = quick_sort(left)[0] + middle + quick_sort(right)[0]
        snapshots.append(sorted_arr.copy())  # store after every partition step
        return sorted_arr, snapshots
```

```python
def visualize_sort(snapshots):
    for step, arr in enumerate(snapshots):
        clear_output(wait=True)
        plt.figure()
        plt.bar(range(len(arr)), arr)
        plt.title(f"Bubble Sort Step {step}")
        plt.xlabel("Index")
        plt.ylabel("Value")
        plt.show()
        time.sleep(0.2)


arr = [64, 34, 25, 12, 22, 11, 90]

sorted_arr, snapshots = insertion_sort(arr)

#visualize_sort(snapshots)

print("Sorted Array:", sorted_arr)
```

6. Heap Sort
</div>

```python
def heap_sort(arr):
    snapshots = [arr.copy()]  # store initial state

    def heapify(arr, n, i):
        largest = i
        l = 2 * i + 1
        r = 2 * i + 2

        if l < n and arr[l] > arr[largest]:
            largest = l

        if r < n and arr[r] > arr[largest]:
            largest = r

        if largest != i:
            arr[i], arr[largest] = arr[largest], arr[i]
            snapshots.append(arr.copy())  # store after every swap
            heapify(arr, n, largest)

    n = len(arr)

    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        snapshots.append(arr.copy())  # store after every swap
        heapify(arr, i, 0)

    return arr, snapshots
```

```python
def visualize_sort(snapshots):
    for step, arr in enumerate(snapshots):
        clear_output(wait=True)
        plt.figure()
        plt.bar(range(len(arr)), arr)
        plt.title(f"Bubble Sort Step {step}")
        plt.xlabel("Index")
        plt.ylabel("Value")
        plt.show()
        time.sleep(0.2)


arr = [64, 34, 25, 12, 22, 11, 90]

sorted_arr, snapshots = insertion_sort(arr)

#visualize_sort(snapshots)

print("Sorted Array:", sorted_arr)
```

7. Shell Sort
</div>

```python
def shell_sort(arr):
    snapshots = [arr.copy()]  # store initial state
    n = len(arr)
    gap = n // 2

    while gap > 0:
        for i in range(gap, n):
            temp = arr[i]
            j = i
            while j >= gap and arr[j - gap] > temp:
                arr[j] = arr[j - gap]
                j -= gap
                snapshots.append(arr.copy())  # store after every shift
            arr[j] = temp
            snapshots.append(arr.copy())  # store after placing the temp
        gap //= 2

    return arr, snapshots
```

```python
def visualize_sort(snapshots):
    for step, arr in enumerate(snapshots):
        clear_output(wait=True)
        plt.figure()
        plt.bar(range(len(arr)), arr)
        plt.title(f"Bubble Sort Step {step}")
        plt.xlabel("Index")
        plt.ylabel("Value")
        plt.show()
        time.sleep(0.2)


arr = [64, 34, 25, 12, 22, 11, 90]

sorted_arr, snapshots = insertion_sort(arr)

#visualize_sort(snapshots)

print("Sorted Array:", sorted_arr)
```
